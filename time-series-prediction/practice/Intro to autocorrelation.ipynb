```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create a chart of now and lag values
def computeCor(temps, shift):
    day_i_temps = temps[shift:]
    day_i_minus_1 = temps[:-shift]
    # shift = 1
    print("series data: \n", temps)
    print("lag 1: \n", day_i_minus_1)    
    print("current: \n", day_i_temps)    
    print("corrcoef: \n", np.corrcoef(day_i_minus_1, day_i_temps))
    plt.subplot(1, 2, 1)
    plt.plot(temps, marker='.')
    plt.ylabel("temperature")
    plt.xlabel("days")
    plt.title('Original Time-Series')
    plt.subplot(1, 2, 2)
    plt.title('Scatter on day-1 as x and day as y')
    plt.scatter(x= day_i_minus_1, y= day_i_temps, )    
    plt.ylabel("day")
    plt.xlabel("day-i")
    plt.show()

# Draw correlation as shift increases
def draw_corcoef(temp3, lag=5):
    autocorrelation = []
    for shift in range(lag):
        shift = shift + 1
        day_i_temps = temps3[shift:]
        day_i_minus_1 = temps3[:-shift]
        #correlation = np.corrcoef(day_i_minus_1, day_i_temps)[0,1]
        correlation = np.corrcoef(day_i_temps, day_i_minus_1)[0,1]
        autocorrelation.append(correlation)
    #print(autocorrelation)
    plt.title("Correlation as lag increases")
    plt.ylabel("corcoef")
    plt.xlabel("lag")
    plt.plot(autocorrelation)
    plt.show()    
```

```python
temps_linear_up = np.array([20, 21, 23, 25, 27])
computeCor(temps_linear_up,1)
draw_corcoef(temps_linear_up)
```

<h1>Partial Correlation</h1>

Start by treating the data as residuals:
left over errors that you haven't been able to fit yet
<br>
The following code means:<br>
We go each day and after we calculate the core coefficient of correlation between that day and the day before. 
We fit the line to it. We calculate the estimate along all the points on that line and then we subtract that estimate from our what we're still trying to fit and we get updated residuals.
So what that means is one day.
we get our original day and one shift and one autocorrelation because the residuals that we're fitting there is actually the original temperatures but after that it falls off very quickly and that means that what we're doing is we're using the termperature two days ago to try to correct for any errors that we missed by using today as an estimate so in the case of temperature once we use the previous day to estimate today's temperature there's actually not alot of information left in the days that comes before. so it falls down to the noise very quickly


```python
temps4 = temps3
pac = []
residuals = temps4
for shift in range(20):
    shift = shift + 1
    print("===Shift: {}======".format(shift))
    day_i = temps4[:-shift]
    res = residuals[shift:]
    #print(len(day_i), len(res))
    print("day_i: ", day_i)
    print("residual: ", res)
    correlation = np.corrcoef(day_i,res)[0,1]
    print("correlation: ", correlation)
    pac.append(correlation)
    
    #Fit the new day's data and find the residuals
    slope, intercept = np.polyfit(
        day_i, res, 1)
    estimate = intercept + slope * day_i
    print("estimate: " , estimate)
    #update residuals
    residuals[shift:] = residuals[shift:] - estimate
    print("residuals[shift:] : ", residuals[shift:])
#print(pac)
plt.plot(pac)
plt.show()
```
