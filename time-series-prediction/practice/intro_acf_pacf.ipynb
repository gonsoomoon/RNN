<h2> Partial Autocorrelation Function</h2><br>
A partial autocorrelation is a summary of the relationship between an observation in a time series with observations at prior time steps with the relationships of intervening observations removed<br>
The PAC of lag k is the correlation that results after removing the effect of any correlations due to the terms at shorter lags

<h2>Moving Average Intuition</h2><br>

Remember that the moving average proces is an autoregression model of the time series of residual errors from prior predictions. Another way to think about the moving average model is that it corrects future forecasts based on errors made on recent forecasts.
<br>
We would expect the ACF for the MA(k) process to show a strong correlation with recent values up to the lag of k, then a sharp decline to low or no correlation.<br>
For the PACF, we would expect the plot to show a strong relationship to the lag and a trailing off of correlation from the lag onwards

```python
from statsmodels.graphics.tsaplots import plot_pacf
series[0:10].plot()
pyplot.show()
plot_pacf(series, lags = 3)
pyplot.show()
```
